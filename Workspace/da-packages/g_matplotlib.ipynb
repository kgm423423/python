```python
#  모듈 준비 
import matplotlib.pyplot as plt
```

```python
# 선 그래프 1
plt.plot([1, 4, 9, 16]) # y축의 값만 적용, x축 값은 자동으로 적용
plt.show() 
```

```python
# 선 그래프 2
plt.plot([10, 20, 30, 40], [1, 4, 9, 16]) # x축 y축 값 적용
plt.show() 
```

```python
# 그래프 스타일 1
plt.plot([1, 4, 9, 16], 'rs--')
plt.show()
```

```python
# 그래프 스타일 2
plt.plot([1, 4, 9, 25], c='b', lw=5, ls="--", marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.show()
```

```python
# 축 설정

plt.plot([1, 4, 9, 25], c='b', lw=5, ls="--", marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(-0.2, 3.2)
plt.ylim(-2, 32)
plt.show()
```

```python
import numpy as np
pi = np.pi

x = np.linspace(-pi, pi, 256) # a ~ b 까지 일정한 간격으로 쪼개기 

c = np.cos(x)
plt.plot(x, c)
plt.xticks([-pi, (-pi/2), 0, (pi/2),  pi], rotation = 90)
plt.yticks([-1, 0, +1])
plt.grid(True)
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
# plt.plot(t, t, 'r--')
# plt.plot(t, 0.5*t**2, 'bs:')
# plt.plot(t, 0.2*t**3, 'g^-')
plt.plot(t, t, 'r--',
         t, 0.5*t**2, 'bs:',
         t, 0.2*t**3, 'g^-') # 위 코드와 동일
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', label=1)
plt.plot(t, 0.5*t**2, 'bs:', label=2)
plt.plot(t, 0.2*t**3, 'g^-', label=3)
plt.legend() # 라벨 보여주기
plt.show()
```

```python
x = np.arange(-5, 10, 2)
y = np.arange(0, 10, 1)
ax1 = plt.subplot(2, 1, 1) # 전체 2행 1열 그래프에서 1번쨰 axes 가져오기
plt.plot(x)
plt.plot(y)
plt.show()
ax2 = plt.subplot(1, 2, 2) # 전체 1행 2열 그래프에서 1번쨰 axes 가져오기
plt.plot(x)
plt.plot(y)
plt.show()

print(ax1, ax2)
```

```python
x1 = np.arange(-5, 10, 2)
y1 = np.arange(0, 10, 1)
fig, axes = plt.subplots(2, 1) 
axes[0].plot(x1)
axes[0].plot(y1)

x2 = np.arange(5, -10, -2)
y2 = np.arange(0, -10, -1)
axes[1].plot(x2)
axes[1].plot(y2)
plt.show()

print(ax1, ax2)
```

```python
labels = ['A', 'B', 'C', 'D']
colors = ['r', 'g', 'b', 'y']
sizes = [15, 30, 10, 45]
explodes = (0, 0.1, 0, 0)
plt.pie(sizes, explode=explodes, labels = labels, colors=colors,
         autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
# 히스토그램 그래프
x = np.random.randn(10000) # 표준정규분포에서 500개의 난수 발생
print(x)

plt.hist(x, bins=100) # 50개의 등급을 나누고 각 등급에 포함된 데이터의 빈도수를 막대그래프로 표시
plt.show()
```

```python
# 산점도 그래프
x = np.random.normal(0, 1, 1024) # 정규분포에서 평균이 0이고 분포도가 1인 1024개의 데이터
y = np.random.normal(0, 1, 1024)
plt.scatter(x, y)
plt.show()
```
