```python
import csv

paths = "../data-files/subway-data-2.csv"
rows = []

with open(paths, 'r', encoding="UTF-8") as f:
    reader = csv.reader(f)

    for row in reader:
        for i in range(4, 14):
            row[i] = row[i].replace(',', '')
        rows.append(row)

print(rows[0])
```

```python
import numpy as np

arr = np.array(rows)
ts = arr[:, 10:14].astype(np.int32).sum(axis=0) # axis=0 행 방향으로 합계 계산

print(f"({ts[0]:,}) + ({ts[1]:,}) + ({ts[2]:,}) + ({ts[3]:,}) = ({ts.sum():,})")
```

```python
# 남성 총인구수, 여성 총인구수
import csv

paths = "../data-files/population.csv"
rows = []

with open(paths, 'r', encoding="UTF-8") as f:
    reader = csv.reader(f)

    print(next(reader))
    for row in reader:
        for i in range(1, len(row)):
            row[i] = row[i].replace(',', '')
        rows.append(row)
```

```python
import numpy as np

# 동읍면만 뽑아서 들고오기
arr = np.array([row for row in rows if row[0][-6] != '0'])

totals = arr[:, [1, 104, 207]].astype('int').sum(axis=0)

print(f"[총 인구수 : {totals[0]:,}]\n[남성 인구수 : {totals[1]:,}]\n[여성 인구수 : {totals[2]:,}]")
```
