```python
# 모듈 준비
import seaborn as sns
import matplotlib.pyplot as plt


tips = sns.load_dataset("tips")
tips.head()
sns.relplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip",hue="smoker", data=tips) # 흡연여부 데이터를 색상에 매핑
plt.show()
```

```python
fmri = sns.load_dataset("fmri")

sns.relplot(x="timepoint", y="signal", hue="region", style="event", kind="line", data=fmri)
plt.show()
```

```python
# 컬럼에 비교할 분류 바인딩
sns.relplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips)
plt.show()
```

```python
# 컬럼에 비교할 분류 바인딩 2
fmri = sns.load_dataset("fmri")

sns.relplot(x="timepoint", y="signal", hue="region",
            col="region", row="event", height=3,
            style="event", kind="line", estimator=None, data=fmri)
```

```python

# frontal_mask = fmri["region"] == "frontal"
# frontal_fmri = fmri[frontal_mask]

# 위 코드와 동일
frontal_fmri = fmri.query("region == 'frontal'") # region 컬럼의 데이터가 'frontal'인 행만 뽑기

sns.relplot(x="timepoint", y="signal", hue="event",
            col="subject", col_wrap=5, height=3, linewidth=2.5,
            style="event", kind="line",  data=frontal_fmri)
```

```python
# 기본 범주 데이터 산점도 그래프
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", hue="day", data=tips, jitter=False)
sns.catplot(x="day", y="total_bill", hue="day" , data=tips, jitter=True) # 오버플로팅 해결 지터링 기법
```

```python
# beeswarm형 그래프 
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="day", kind="box", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
plt.show()
```

```python
# 단일 바이올린 플롯
sns.catplot(x="day", y="total_bill", hue="sex", kind="violin", split=True, data=tips)
```

```python
# 바이올린 플롯과 범주데이터 산점도 그래프 병합
sns.catplot(x="day", y="total_bill", kind="violin", hue="sex", inner=None, split=True, data=tips)
sns.swarmplot(x="day", y="total_bill", color='k', size=3, data=tips)
plt.show()
```

```python
# Bar plot
titanic = sns.load_dataset("titanic")

print(titanic.groupby(by=["sex"])["alive"].count())

sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
titanic.head()
```

```python
sns.catplot(x="class", y="survived", hue="sex", kind="point", data=titanic)
plt.show()
```

```python
exercise = sns.load_dataset("exercise")
sns.catplot(x="time", y="pulse", hue="kind", col="diet", data=exercise)
plt.show
```

```python
# 일반량 분포 시각화
import numpy as np

x = np.random.normal(size=10000)
sns.distplot(x, bins=20, kde=True,rug=True) # kde 추세선,  rug 실 위치 표시
plt.show()
sns.distplot(x, bins=20, kde=False,rug=True)
plt.show()
```

```python
iris = sns.load_dataset("iris")

sns.pairplot(iris, hue="species", markers=['o', 's', '^'])
plt.show()
```

```python
# 선형 관계 시각화
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```
