```python
# 데이터분석 패키지 준비
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 준비 : pandas를 사용해서 titanic.csv 읽기

paths = "../data-files/titanic.csv"

titanic_df = pd.read_csv(paths)
```

```python
# 데이터 확인 : info, shape, head, tail

print(titanic_df.info())
print(titanic_df.shape)
print(titanic_df.head(3))
print(titanic_df.tail(3))
```

```python
# 성별 생존률 계산 : grouphby
print(titanic_df.groupby(by=["Sex"])["Survived"].sum())
print(titanic_df.groupby(by=["Sex"])["Survived"].count())

# 성별 생존율 시각화
sns.catplot(x="Sex", y="Survived", kind="bar", data=titanic_df)
plt.show()
```

```python
# 승선 등급, 가족동행여부, 승선항, 
sns.catplot(x="Pclass", y="Survived", hue="Pclass", kind="bar", data=titanic_df)
plt.show()
# print(titanic_df.groupby(by=["SibSp"])["Survived"].count())
# print(titanic_df.groupby(by=["SibSp"])["Survived"].sum())
#l1 = titanic_df.groupby(by=["SibSp"])["Survived"].sum() / titanic_df.groupby(by=["SibSp"])["Survived"].count()
#plt.bar([i for i in range(l1.size)], l1)
adult_titanic_df = titanic_df[titanic_df["Age"] > 20]
sns.catplot(x="SibSp", y="Survived", hue="SibSp",col="Sex", kind="bar", data=adult_titanic_df)
sns.catplot(x="Parch", y="Survived", hue="Parch",col="Sex", kind="bar", data=adult_titanic_df)
sns.catplot(x="Embarked", y="Survived", hue="Embarked", kind="bar",estimator="sum", data=titanic_df)
```
