```python
# subway-data-2.csv 파일의 데이터 읽어서 데이터프레임에 저장
# 이용객수가 가장 많은 지하철역 찾기 4 ~ -2 까지 합
# iloc 함수를 사용하여 계산 df.iloc[:, 10:20] 모든 행에서 10 에서 20번쨰 까지 컬럼
import numpy as np
import pandas as pd

paths = "../data-files/subway-data-2.csv"

#  thousands=',' 데이터에 포함된 쉼표를 천단위 구분자로 해석
subway_df = pd.read_csv(paths, header=None, thousands=',')


# 각 행별 이용객수의 합을 계산해서 tot 컬럼에 저장
subway_df["tot"] = subway_df.iloc[:, 4:-2].sum(axis=1) # axis=0 열별 합계, axis=1 행별 합계 
sorted_subway_df= subway_df.sort_values("tot", ascending=False)
sorted_subway_df[[0,1,2,3,"tot"]].head(10)


```

```python
# population.csv 파일의 데이터 읽어서 데이터프레임에 저장
# 읍면동 수준에서 인구가 갖아 많은 행정구역, 가장 적은 행정구역 찾기

paths = "../data-files/population.csv"

pop_df = pd.read_csv(paths, encoding="UTF-8", thousands=',')

dong_mask = pop_df["행정구역"].str.get(-6) != '0' # 뒤에서부터 6번쨰 문자 뽑기 

pop_by_dong_df = pop_df[dong_mask]
sorted_df = pop_by_dong_df.sort_values("2025년04월_계_총인구수", ascending=False)
print(sorted_df.iloc[:, :2].head())
print(sorted_df.iloc[:, :2].tail())
```

```python
pd.read_csv("../data-files/population.csv", encoding="UTF-8", thousands=',')[pop_df["행정구역"].str.get(-6) != '0'].sort_values("2025년04월_계_총인구수", ascending=False).iloc[:, :2].head()
```

```python
# 남성 인구비가 가장 높은 읍면동 찾기 

male_dong_df = pop_by_dong_df
male_dong_df["male_ratio"] = pop_by_dong_df["2025년04월_남_총인구수"] / pop_by_dong_df["2025년04월_계_총인구수"]
male_dong_df.sort_values("male_ratio", ascending=False).iloc[:,[0,-1]].head()
```

```python
# 각 연령대별 인구 컬럼 만들기
# 20대 인구비가 가장 높/낮은 읍면동 찾기
pop_df2 = pd.DataFrame()
pop_df2["행정구역"] = pop_df["행정구역"]
pop_df2["총인구수"] = pop_df["2025년04월_계_총인구수"] 
for i in range(0, 100, 10):
    pop_df2[str(i)+"대_인구수"] = pop_df.iloc[:, i+4:i+14].sum(axis=1)

dong_df2 = pop_df2[pop_df2["행정구역"].str.get(-6) != '0']

sorted_df2 = pd.DataFrame()
sorted_df2["행정구역"] = dong_df2["행정구역"]
sorted_df2["20대_인구비"] = (dong_df2["20대_인구수"] / dong_df2["총인구수"])
sorted_df2.sort_values("20대_인구비", ascending=False)

print(sorted_df2[["행정구역","20대_인구비"]].head(5))
print("\n")
print(sorted_df2[["행정구역","20대_인구비"]].tail(5))
```
