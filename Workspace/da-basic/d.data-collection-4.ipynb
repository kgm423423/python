```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import seaborn as sns
```

```python
response = requests.get('https://namu.wiki/w/%EA%B5%AD%EA%B0%80%EB%B3%84%201%EC%9D%B8%EB%8B%B9%20%EB%AA%85%EB%AA%A9%20GDP%20%EC%88%9C%EC%9C%84')
html = response.content.decode('utf-8')
soup = BeautifulSoup(html, 'html.parser')
table = soup.find('table')
```

```python
data = []
for indx, tr in enumerate(table.find_all('tr')):

    tds = indx == 0 if tr.select('th') else tr.select('td')
    data.append([td.get_text(strip=True) for td in tds])

df = pd.DataFrame(data[1:], columns=data[0])
df
```
