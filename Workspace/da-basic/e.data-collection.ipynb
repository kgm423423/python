```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
data=[]

for i in range(52):
    url = f"https://korean.visitseoul.net/attractions?curPage={i+1}"
    response = requests.get(url)
    http = response.content.decode("utf-8")
    soup = BeautifulSoup(http, "html.parser")
    titles = soup.select('span[class="title"]')
    for title in titles:
        data.append(title.text.strip())
        print(title.text.strip())
    time.sleep(0.2) 




```

```python
df = pd.DataFrame(data, columns=["title"])
df
```
