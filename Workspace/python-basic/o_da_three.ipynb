```python
# 특정 읍면동의 인구 구성을 그래프로 보기

import csv

paths = "data-files/subway-data-1.csv"
rows = []

# 1 파일 열기
with open(paths, 'r', encoding="UTF-8") as f:
    reader = csv.reader(f)
    print(next(reader)) # 컬럼 제거

    # 쉼표 제거 + 정수형 변환
    for row in reader:  
        try :
            for i in range(4,8):
                row[i] = int(row[i].replace(',', ''))
            row[2] = int(row[2])
        # 예외
        except Exception as e:  print("error", e)

        rows.append(row)

# # 출력
# for row in rows:
#     for v in row:
#         print(v,end=" ")
#     print()

# 이용객 수가 가장 많은 역 찾기

# maxStation = None

# for row in rows:
#     if (maxStation == None or sum(maxStation[4:7]) < sum(row[4:7])):
#         maxStation = row
    
# print(maxStation[3], sum(maxStation[4:7]))

# 튜플로 만들어서 2번쨰 인자로 내림차순 정렬
stationList = []

for row in rows:
    stationList.append((row[3], sum(row[4:7]), row[1]))

stationList.sort(key = lambda x : -x[1])

for i in range(len(stationList)):
    print("{0:>3}: {1}".format(str(i), stationList[i]))

# 막대 그래프 그리기
import matplotlib.pyplot as plt

plt.bar(range(len(stationList)), [v[1] for v in stationList], color="green")

plt.show()







```

```python
# subway-data-2.csv 파일의 데이터를 읽어서 리스트로 구성
# csv모듈 사용, 숫자형으로 변환, 제목 행 없음

paths = "data-files/subway-data-2.csv"
rows2 = []

# 1 파일 불러오기
with open(paths, 'r', encoding="UTF-8") as f:
    reader = csv.reader(f)
    for row in reader:
        rows2.append(row)


# 2 쉼표 제거하고 형변환 
for row in rows2:
    row[2] = int(row[2].lstrip('0'))
    for i in range(4, len(row)-1):
        row[i] = int(row[i].replace(',', ''))

# test
for row in rows:
    for i in range(len(row)):
        print(row[i], end=" ")
    print()

    
```
