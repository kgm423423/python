```python
# population.csv 파일을 읽어서 리스트에 변수를 저장

# 1 파일 불러오기
paths = 'data-files/population.csv'

lines = []
with open(paths, 'r', encoding="UTF-8") as f:
    lines = f.readlines()

# 2 파일 잘라서 리스트로 만들기
rows = []
for line in lines[1:10]:
    row = []
    i = 0
    j = 0
    while (line[i] != '\n'):
        while (line[i] != '\"'): i += 1
        j=i+1
        while (line[j] != '\"'): j += 1
        row.append(line[i+1:j])
        i = j+1
    rows.append(row)

# 3 문자열 숫자로 변환 
for row in rows:
    for value in row[1:]:
        try:
            value = value.replace(",","")
            value = int(value)
        except: print("error")



    
        
```
