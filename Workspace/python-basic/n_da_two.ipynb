```python
# population.csv 파일을 읽어서 리스트에 변수를 저장

# 1 파일 불러오기
paths = 'data-files/population.csv'

lines = []
with open(paths, 'r', encoding="UTF-8") as f:
    lines = f.readlines()

# 2 파일 잘라서 리스트로 만들기
rows = []
for line in lines[1:]:
    row = []
    i = 0
    j = 0
    while (line[i] != '\n'):
        while (line[i] != '\"'): i += 1
        j=i+1
        while (line[j] != '\"'): j += 1
        row.append(line[i+1:j])
        i = j+1
    rows.append(row)
# rows = []
# for line in lines[1:]:
#     row = line.split('","')
#     row[0] = row[0][1:]
#     row[len(row)-1] = row[len(row)-1][:-2]
#     rows.append(row)



# 3 문자열 숫자로 변환 
# for row in rows:
#     for value in row[1:]:
#         try:
#             value = value.replace(",", "")
#             value = int(value)
#         except: print("error")

for row in rows:
    for i in range(1, len(row)):
        try:    row[i] = int(row[i].replace(",", ""))
        except: print("error")    
    
print(rows[len(rows)-1])
```

```python
# scv 모듈을 사용해서 csv 파일 읽기

import csv

paths = 'data-files/population.csv'
rows2 = []
titles = []

with open(paths, 'rt', encoding="UTF-8") as f:
    reader = csv.reader(f) # csv 모듈을 사용해서 파일 읽기 준비 

    titles = next(reader) # 현재 행을 반환하고 다음 행으로 이동
    for row in reader:    # 파일을 다 읽을 때까지 next를 반복
        for i in range(1, len(row)): 
            row[i] = int(row[i].replace(",", ""))
        rows2.append(row)

```

```python
# 읍면동 수준에서 인구가 가장 많은 구역 찾기
# 읍면동 수준은 각 행의 첫 번쨰 데이터(행정구역)의 뒤에서 6번째 문자가 0이 아닌 경우

max_pop = None
min_pop = None

for row in rows2:
    if ( row[0][-6] != '0'):
        if ((max_pop == None) or (max_pop[1] < row[1])):
            max_pop = row
        if ((min_pop == None) or (min_pop[1] > row[1])):
            min_pop = row
            
print(max_pop)
print(min_pop)
    

```

```python
# 남성 인구비가 가장 높은 / 낮은 읍면동 찾기 
# 종인구(1번) 남성인구(104번)

max_male_pop = None
min_male_pop = None

def f(rowf):
    try:    return (rowf[104] / rowf[1])
    except: return -1

for row in rows2:
    if (row[0][-6] != '0'):
        rowP = f(row)
        maxP = f(max_male_pop)
        minP = f(min_male_pop)

        if (max_male_pop == None or maxP != -1 and rowP != -1 and maxP < rowP): 
            max_male_pop = row
        if (min_male_pop == None or minP != -1 and rowP != -1 and minP > rowP): 
            min_male_pop = row

print(f(max_male_pop), max_male_pop)
print(f(min_male_pop), min_male_pop)


        
```
